```python
import pandas as pd
import psycopg2
```

```python
connection = psycopg2.connect(user="meditreats",
                              password="meditreats",
                              host="df-treats-db.cs6hxh6ocizm.us-west-2.rds.amazonaws.com",
                              port="5432",
                              database="meditreats")
cursor = connection.cursor()

```

```python
query = "select * from effectsgroups where study is NULL"

cursor.execute(query)

effects_groups_without_a_study = cursor.fetchall()

effects_groups_without_a_study = pd.DataFrame(effects_groups_without_a_study, columns=['id', 'title', 'description', 'study_id', 'study'])

effects_groups_without_a_study
```

```python
effects_groups_without_a_study = effects_groups_without_a_study[effects_groups_without_a_study["title"].str.contains("lacebo")==False]

effects_groups_without_a_study = effects_groups_without_a_study.rename(columns={"study_id": "group_id", "id": "db_id"})

effects_groups_without_a_study_cleaned = effects_groups_without_a_study.drop(columns=['study'])

effects_groups_without_a_study_cleaned
```

```python
effects_groups_data = pd.read_pickle('/Users/pauldittamo/Downloads/effects_groups_treats.pkl')

effects_groups_data = effects_groups_data.drop(columns=['desc_plus_title', 'treatments', 'adjusted'])

effects_groups_data
```

```python
effects_groups_added_studies = effects_groups_without_a_study_cleaned.merge(effects_groups_data, on=['title', 'description', 'group_id'], how='left')


effects_groups_added_studies
```

```python
removed_duplicates = effects_groups_added_studies[effects_groups_added_studies[['db_id']].duplicated() == False]

removed_duplicates
```

```python
cleaned_effects_groups = removed_duplicates.rename(columns={"study_id":"study", "group_id":"study_id"})


cleaned_effects_groups
```

```python
# temp = cleaned_effects_groups.head(100)

rows = zip(cleaned_effects_groups.db_id, cleaned_effects_groups.study)
```

```python
cursor.execute("""CREATE TEMP TABLE studies_temp(id INTEGER, study VARCHAR ( 255 )) ON COMMIT DROP""")
cursor.executemany("""INSERT INTO studies_temp (id, study) VALUES(%s, %s)""", rows)

cursor.execute("""
    UPDATE effectsgroups
    SET study = studies_temp.study
    FROM studies_temp
    WHERE studies_temp.id = effectsgroups.id;
    """)

connection.commit()
```
